```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 1001)
y = np.sin(x)
```

```python
initial_point = 300
window_size = 200
look_ahead = 20
plt.plot(y)
plt.plot([None]*initial_point + list(y[initial_point:initial_point+window_size]), color='r')
plt.scatter(initial_point+window_size+look_ahead, np.sin(x[initial_point+window_size+look_ahead]),
            c='red', marker='*')
```

```python
import pandas as pd
```

```python
df = pd.read_csv('monthly_csv.csv')
```

```python
df = df[df['Source'] == 'GCAG'][::-1]
```

```python
df = df.reset_index().drop(['index', 'Source'], axis=1)
```

```python
df.plot(figsize=(20, 8))
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.set_index('Date')
```

```python
df.plot(figsize=(20, 8))
```

```python
window = 30
look_ahead = 10
X = []
Y = []
for i in range(df.shape[0]-window-look_ahead):
    X.append(df.iloc[i:i+window].values)
    Y.append(df.iloc[i+window+look_ahead].values)
    
X = np.array(X)
Y = np.array(Y)
```
